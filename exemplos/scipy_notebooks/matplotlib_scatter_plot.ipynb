```python
from matplotlib import pyplot as plt
```

```python
T= [0.0, 1.1, 2.4, 3.2, 4.8, 5.2, 6.0]
P= [1.0, 1.1, 1.0, 1.5, 1.6, 1.1, 2.1]
```

```python
plt.scatter(T,P)
plt.show()
```

```python
x=list(range(100))
from math import sin
y=[sin(xi/10) for xi in x]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(T,P,marker='*',color='y',label='amostra 1')
plt.plot(x,y,ls=':',color='b', label='modelo oscilat√≥rio')
plt.title("seno estrelado")
plt.ylabel("eixo y")
plt.xlabel("eixo x")
plt.xlim(0,150)
plt.ylim(-.5,2.5)
plt.legend()
plt.savefig("Figura1.png")
plt.show()
```
