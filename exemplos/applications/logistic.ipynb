```python
import numpy as np
```

```python
%%timeit
n=1000
y=n*[0.]

k=3.912854
y[0]=.341238563

for i in range(1,n):
    y[i]=k*y[i-1]*(1.-y[i-1])
```

# lists

```python
%%timeit
n=100

k=3.912854
y=[.341238563]

for i in range(1,n):
    y.append(k*y[i-1]*(1.-y[i-1]))
```

```python
from numba import jit

@jit
def logistic():
    n=100
    y=np.zeros(n)

    k=3.912854
    y[0]=.341238563

    for i in range(1,n):
        y[i]=k*y[i-1]*(1.-y[i-1])
    return y
```

```python
y=logistic()
```

```python
%%timeit
y=logistic()
```

```python
from numba import jit

@jit
def logistic_L():
    n=100

    k=3.912854
    y=[.341238563]

    for i in range(1,n):
        y.append(k*y[i-1]*(1.-y[i-1]))
    return y
```

```python
y=logistic_L()
```

```python
%%timeit
y=logistic_L()
```

```python
from numba import jit

@jit
def logistic_L2():
    n=100
    y=100*[0.]
    k=3.912854
    y[0]=.341238563

    for i in range(1,n):
        y[i]=k*y[i-1]*(1.-y[i-1])
    return y
```

```python
y=logistic_L2()
```

```python
%%timeit
y=logistic_L2()
```
